```python
try:
    3/0
except Exception as e:
    print(e)# here we are just catching the error and printing it
```

```python
#best practice for above code
try:
    3/0
except ZeroDivisionError as e:# best practice is specifying the exact error in exception handling
    print(e)
```

```python
# the advantage of specifying the exception is we can handle it like if i know there will a zero division error i can modify it 
try:
    3/'3'
except TypeError as e:
    print(3/int('3'))
    print(e)
```

```python
#handling two specific errors at the same time
try:
    2/'sfd'
except (TypeError,ZeroDivisionError) as e:
    print('div is not possible ',e)
```

```python
#we can handle multiple errors in blocks
try:
    2/0
except TypeError as e:
    print('div is not possible ',e)
except ZeroDivisionError as e:
    print('div is not possible ',e)
```

```python
#since we might know the possible exceptions so we use parent exception class as best practice as well
#example >> we know in div zero div error may occur and dont know of other combos so we place the known error handling first and then used the parent exception handling to handle the unknown errors
try:
    2/0
except ZeroDivisionError as e:
    print('div is not possible ',e)
except Exception as e:
    print(e)
```
