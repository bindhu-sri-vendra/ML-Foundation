```python
class ValidateSalary(Exception): #here exception is the parent class which holds all exceptions
    def __init__(self,msg):
        self.msg=msg
```

```python
#custom exception
def validate_salary(salary):
    if salary<=0:
        raise ValidateSalary('salary cant be negative')
    elif salary>=300000:
        raise ValidateSalary('salary is too much')
```

```python
try:
    salary=int(input('enter your salary'))
    validate_salary(salary)
except ValidateSalary as e:
    print(e)
else:
    print('salary is valid')
```
