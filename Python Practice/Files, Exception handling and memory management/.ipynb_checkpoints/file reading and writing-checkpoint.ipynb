```python
text_file="""
sample one
sample two 
sample three """
```

```python
#to write a text file
with open('sample.txt','w')as txt:
    txt.write(text_file)
```

```python
#to read a text file
with open('sample.txt','r')as tt:
    data=tt.read()
print(data)
```

```python
#comma separated values >> csv files
data=[['name','course','fee'],['bindhu','DS','2000'],['tara','QA','5000']]
```

```python
#to write a csv file
import csv
```

```python
with open('example_csv.csv','w')as f:
    w= csv.writer(f)
    for i in data:
        w.writerow(i)
```

```python
#to read from a csv file
with open('example_csv.csv','r')as f:
    r=csv.reader(f)
    for i in r:
        print(i)
```

```python
#JSON >> javascript object notation file
dat={'name':'bindhu','uni':'kaist','fee':4000}
```

```python
type(dat)
```

```python
#to write in a json file
import json
with open('json_file.json','w')as f:
    json.dump(dat,f)
```

```python
#to read from a json file
with open('json_file.json','r')as f:
    d=json.load(f)
print(d)
```

```python
#to write a binary file
with open('exam_bin.bin','wb')as f:#we use wb mode which means write binary
    f.write(b'\x68\x65\x6c\x6c\x6f')
```

```python
#to read a binary file
with open('exam_bin.bin','rb') as f:
    print(f.read())
```

```python
#buffered reading and writing
#it allows us to read or write data in chunks
import io
with open('buff.txt','wb')as f:
    file=io.BufferedWriter(f)
    file.write(b'A portion, piece, or segment that is representative of a whole.showed samples of a new stretch fabric')
    file.write(b'hello bindhu')
    file.write(b'congratulations on getting into kaist')
    file.flush()
```

```python
with open('buff.txt','rb')as f:
    file=io.BufferedReader(f)
    data=file.read(25) #here we can mention how many bytes we want to read 
    print(data)
```
