```python
# to get the execution time of the program
import time
start = time.perf_counter()
def test_fun():
    print('do something')
    print('sleeping for 1 sec')
    time.sleep(1)
    print('done sleeping')
test_fun()
test_fun()
test_fun()
test_fun()
test_fun()
end = time.perf_counter()
print('the program is finished in ', round(end-start,2))
```

```python
import time
import threading
start = time.perf_counter()
def test_fun():
    print('do something')
    print('sleeping for 1 sec')
    time.sleep(1)
    print('done sleeping')

t1=threading.Thread(target=test_fun)
t2=threading.Thread(target=test_fun)
t1.start()
t2.start()
t1.join()
t2.join()# join makes sure that t1 and t2 executes first and then main thread is executed
end = time.perf_counter()
print('the program is finished in ', round(end-start,2))
```

```python
#using thereads with loops
import time
import threading
start = time.perf_counter()
def test_fun():
    print('do something')
    print('sleeping for 1 sec')
    time.sleep(1)
    print('done sleeping')
threads = []
for i in range(5):#we gonna run 5 threads
    t=threading.Thread(target=test_fun)
    t.start()
    threads.append(t)

for i in threads:
    i.join()
end = time.perf_counter()
print('the program is finished in ', round(end-start,2))
```

```python
#using thereads with functions with arguments
import time
import threading
start = time.perf_counter()
def test_fun(sec):
    print('do something')
    print('sleeping for ',sec,' sec')
    time.sleep(sec)
    print('done sleeping')
threads = []
for i in range(5):#we gonna run 5 threads
    t=threading.Thread(target=test_fun,args=[2])
    t.start()
    threads.append(t)

for i in threads:
    i.join()
end = time.perf_counter()
print('the program is finished in ', round(end-start,2))
```

```python
#shared variable accorss the threads
start=time.perf_counter()
shared_counter=0
counter_lock=threading.Lock()# it basically locks the counter for a specific thread
def increment_shared_counter(x):
    global shared_counter
    with counter_lock:
        shared_counter=shared_counter+1
        print(f'Thread{x}: is incremented to{shared_counter}')
        time.sleep(1)
thread=[threading.Thread(target= increment_shared_counter,args=(i,)) for i in range(5)]
for i in thread:
    i.start()
for i in thread:
    i.join()
end = time.perf_counter()
print('the program is finished in ', round(end-start,2))
```
