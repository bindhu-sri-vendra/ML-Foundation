```python
#function decorators
def fun():
    print('before computation')
    print(34*56)
    print('after computation')
fun()
```

```python
#example of function decorator
def my_decorator(func):
    def wrapper():
        print('before computation')
        func()
        print('after computation')
    return wrapper
```

```python
@my_decorator#here name can be anything we just need to say @decorator function name
def example():
    print('Hey Bindhu')
```

```python
example()
```

```python
#class decorator
class dec:
    def __init__(self,func):
        self.func=func
    def __call__(self):# it is a special function that invokes the instance of the class
        print('before')
        self.func()
        print('after')
```

```python
@dec
def greet():
    print('hey')
greet()
```

```python
# showing the call function activity
class dect:
    def __init__(self):
        print('we are inside the init function')
    def __call__(self):# it is a special function that invokes the instance of the class
        print('before')
```

```python
ob=dect()
```

```python
ob() # calling the instance of the class will invoke the call function
```

```python
#classmethod
class A:
    @classmethod
    def add(self,x,y):
        return x+y
```

```python
A.add(4,6)
```

```python
#staticmethod -- used when we dont wanna interact with class level data 
class A:
    @staticmethod
    def add(x,y): #no need to add cls or self for static method
        return x+y
```

```python
A.add(4,6)
```

```python
#property decorator -- it allows method to be accessed as an attribute
class Circle:
    def __init__(self,radius):
        self.radius=radius
    @property
    def area(self):
        rad=self.radius
        return 3.14*rad**2
```

```python
ob=Circle(5)
ob.area #no need to pass the radius value
```
