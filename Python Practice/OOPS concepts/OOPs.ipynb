```python
#extract even and odd numbers from a list
class Lis:
    l=[1,2,3,4,5,6]
    def extract_even(self,l1):
        ans=[]
        for i in l1:
            if i%2 ==0:
                ans.append(i)
        return ans
    def extract_odd(self,l1):
        ans=[]
        for i in l1:
            if i%2 !=0:
                ans.append(i)
        return ans
a=Lis()
a.l
```

```python
a.extract_even(a.l)
```

```python
a.extract_odd([2,3,4,22,34,53,54,56,66,57])
```

```python
class Lis:
    def __init__(self,l):# init is basically initialization and it stores a unique valure for each object 
        self.l=l
        
    def extract_even(self,l1):
        ans=[]
        for i in l1:
            if i%2 ==0:
                ans.append(i)
        return ans
    def extract_odd(self,l1):
        ans=[]
        for i in l1:
            if i%2 !=0:
                ans.append(i)
        return ans
```

```python
a=Lis([1,2,3,4,5])
a.extract_odd(a.l)
```
