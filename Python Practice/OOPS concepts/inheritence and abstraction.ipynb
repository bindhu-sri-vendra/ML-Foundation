```python
#single inheritence - one parent and one child class --- parent to child 
class Father:
    def fatherproperty(self):
        print('father has a property')
class Son(Father):
    def job(self):
        print('son has a property from job')
```

```python
child_ob=Son()
child_ob.fatherproperty()
child_ob.job() #child objects can use methods from their class and their parent class as well
# the parent class can only access their methods not their child class methods
parent_ob = Father()
parent_ob.fatherproperty()
```

```python
#method overriding
class Fruit:
    def fruit_info(self):
        print('This is parent class')
class Apple(Fruit):
    def fruit_info(self):
        print('This is child class')
    
```

```python
ob=Apple()
ob.fruit_info()#child class method overrides parent class methods --- child class is more powerful
```

```python
#multilevel inheritence ---- parent to child1 to child2
class Grandpa:
    def gram(self):
        print('parent class')
class Papa(Grandpa):
    def father(self):
        print('child one class')
class Son(Papa):
    def child(self):
        print('child two class')
```

```python
ob=Son()
ob.child()
ob.father()
ob.gram()
```

```python
ob1=Papa()
ob1.father()
ob1.gram()
```

```python
#multiple inheritence -- one child inherit the properties of multiple parents
class Parent1:
    def fun1(self):
        print('parent1 class')
class Parent2:
    def fun2(self):
        print('parent2 class')
class Parent3:
    def fun3(self):
        print('parent3 class')
class Child(Parent1,Parent2,Parent3):
    def fun4(self):
        print('This is child class')
```

```python
a=Child()
a.fun1()
a.fun2()
a.fun3()
a.fun4()
```

```python
b=Parent2()
b.fun2() #parent classes can only access their own methods 
```

```python
#diamond problem -- here both parents have same method so there is ambiguitiy
# to resolve it python uses method resolution order(MRO) called c3 linearization so basically the method thats inherited first will be executed 
#in this case parent 2 is inherited first so it will execute
class Parent2:
    def fun2(self):
        print('parent2 class')
class Parent3:
    def fun2(self):
        print('parent3 class')
class Child(Parent2,Parent3):
    def fun4(self):
        print('This is child class')
```

```python
a=Child()
a.fun2()
```

```python
#hierarical inheritence --- one parent more child classes will inherit that one parent
class Mama:
    def ma(self):
        print('this is parent class')
class child1(Mama):
    def fun1(self):
        print('this is first child')
class child2(Mama):
    def fun1(self):
        print('this is second child')
```

```python
ob1=child1()
ob1.fun1()
ob1.ma()
```

```python
#hybrid inheritence --- combination of any inheritence types --- a to b and c, b and c to d
class A:
    def ab(self):
        print(' this is class a')
class B(A):
    def bc(self):
        print(' this is class a')
class C(A):
    def cd(self):
        print(' this is class a')
class D(B,C):
    def de(self):
        print(' this is class a')
```
