```python
#useful dunder methods
#__init__
class Student:
    def __init__(self):# this init mthod is basically used to take data with respective to the object
        print('this will get executed when an instance is created')
ob1=Student()
ob2=Student()
```

```python
#__new__ -- it will be executed before the init and it is responsible to create a new instance of class
class Student:
    def __new__(cls):
        print('this will get executed first')
    def __init__(self):
        print('this will get executed when an instance is created')
ob1=Student()
```
