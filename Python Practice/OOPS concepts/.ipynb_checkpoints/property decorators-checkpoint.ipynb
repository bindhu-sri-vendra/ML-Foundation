```python
class Student:
    def __init__(self, name , price):
        self.__name=name #that double underscore makes it private vairable
        self.__price=price
```

```python
s=Student('bindhu',500)
```

```python
#another way to expose private variables
class Student:
    def __init__(self, name , price):
        self.__name=name 
        self.__price=price
    @property
    def access_price(self):
        return self.__price
```

```python
a=Student('bindhu',5099)
a.access_price
```

```python
#to set the price
class Student:
    def __init__(self, name , price):
        self.__name=name 
        self.__price=price
    @property
    def access_price(self):
        return self.__price
    @access_price.setter #we use this
    def set_price (self,new_price):
        self.__price=new_price
    
```

```python
a=Student('bindhu',5099)
a.set_price=5000
```

```python
a.access_price
```

```python
#to delete the variable we use this 
@access_price.deleter
def del_price(self):
    del self.__price
```
