```python
class Teacher:
    def lecinfo(self):
        print('this is lec info from teacher prespective')
class Student:
    def lecinfo(self):
        print('this is lec info from student prespective')
```

```python
ob1=Teacher()
ob2=Student()
class_ob=[ob1,ob2]
```

```python
def classinfo(class_ob):
    for i in class_ob:
        i.lecinfo()
```

```python
classinfo(class_ob)
```

```python
#method overloading -- same method name with different set of arguments -- it happens in the same class
class Student:
    def f1(self):
        print('welcome to kaist')
    def f1(self,name=""):
        print('welcome to kaist ',name)
    def f1(self,name="",course=""):
        print('welcome to kaist ',name,course)
```

```python
a=Student()
```

```python
a.f1()
```

```python
a.f1('bindhu')
a.f1('bindhu','graduate school of data science')
```

```python
#in python we can write method overloading like this
class Example:
    def fun(self,name='',course=''):
        print('welcome to kaist',name,course)
```

```python
ob=Example()
ob.fun()
```

```python
ob.fun('bindhu')
```

```python
ob.fun('bindhu','gsds')
```
