```python
class Teacher:
    def lecinfo(self):
        print('this is lec info from teacher prespective')
class Student:
    def lecinfo(self):
        print('this is lec info from student prespective')
```

```python
ob1=Teacher()
ob2=Student()
class_ob=[ob1,ob2]
```

```python
def classinfo(class_ob):
    for i in class_ob:
        i.lecinfo()
```

```python
classinfo(class_ob)
```

```python
#method overloading -- same method name with different set of arguments -- it happens in the same class
class Student:
    def f1(self):
        print('welcome to kaist')
    def f1(self,name=""):
        print('welcome to kaist ',name)
    def f1(self,name="",course=""):
        print('welcome to kaist ',name,course)
```

```python
a=Student()
```

```python
a.f1()
```

```python
a.f1('bindhu')
a.f1('bindhu','graduate school of data science')
```

```python
#example of a public data 
#public data can be accessed inside and outside out of the class
class University:
    def __init__(self,name='',course=''):
        self.name=name
        self.course=course
```

```python
ob=University('bindhu','data science')
ob.name
```

```python
ob.name='tarajit'
ob.name# so basically any one can change stuff cause this is a public data
```

```python
#accessing the public method from another method from inside the class
class University:
    def __init__(self,name='',course=''):
        self.name=name
        self.course=course
    def show(self):
        print('name is ',self.name,'course is ',self.course)#this is where we are accessing
```

```python
ob=University('bindhu','data science')
ob.show()
```

```python
#private data can only be accessed inside the class
# we can make data private by adding two underscores before its name --- def __funcname()
class University:
    def __init__(self,name='',course=''):
        self.name=name
        self.__course=course
    def show(self):
        print('name is ',self.name,'course is ',self.__course)
```

```python
ob=University('bindhu','data science')
ob.name
```

```python
ob.course #error cause we made it private
```

```python
#to access the data we need to go through the class
ob._University__course
```

```python
#creating a private method
class University:
    def __init__(self,name='',course=''):
        self.name=name
        self.__course=course
    def show(self):
        print('name is ',self.name,'course is ',self.__course)
    def __privatemethod(self):
        print('This is a private method')
```

```python
ob=University('bindhu','data science')
```

```python
ob.show()
```

```python
ob._University__privatemethod() #to access a private method
```

```python
#use case for private method
class bank:
    def __init__(self,amount):
        self.__amount=amount
    def deposit(self,amount):
        self.__amount+=amount
        print(str(amount),'is deposited')
    def showbalance(self):
        print('your balace is ',self.__amount)
```

```python
cust=bank(1000)
cust.deposit(2000)
```

```python
cust.showbalance()
```

```python
#protected can be used for class and subclass and is made using single underscore
#accessing var from base class --- classname__init__(self)
class A:
    def __init__(self):
        self._collegename='kaist'
class B(A):
    def info(self):
        #A.__init__(self) one way to access and other is to use super class
        super().__init__()
        print('bindhu is studying at ',self._collegename)
```

```python
ob=B()
ob.info()
```
