```python
#usage of queue
from collections import deque
checkout=deque()
checkout.append('cust1')
checkout.append('cust2')
checkout.append('cust3')
checkout.append('cust4')
checkout
```

```python
#to remove an element we use popleft in queue cause first ele should be remove
checkout.popleft()
```
