```python
def example(l):
    a=[]
    for i in l:
        if isinstance(i,int):
            a.append(i)
    return a
example([3,5,'abc','e',6,2])
            
```

```python
#this function works for nested lists as well
def example(l):
    a=[]
    for i in l:
        if type(i)==list:
            for j in i:
                if type(j)==int:
                    a.append(j)
        elif type(i)==int:
            a.append(i)
    return a
example([3,5,'abc','e',6,2,[3,6,7,'kfnkd']])
```

```python
#variable length arguments when we dont know how many arguments will be there then we can use this
def den(*arg):
    sum=0
    for i in arg:
        sum+=i
    return sum
den(3,5)
```

```python
den(3,5,4,7,5,7,9)
```

```python
# variable argument with default argument
def dd(*args,a=0):
    return a
dd(2,3,5,a)
```

```python
#function that takes any no of values but returns a list
def ll(*arg):
    l=[]
    for i in arg:
        if type (i)==list:
            l.append(i)
    return l
ll([2,3,4],5,6,'dfjdhf',[3,5,4])
```
