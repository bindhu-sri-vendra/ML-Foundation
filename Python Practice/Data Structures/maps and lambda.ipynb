```python
#syntax of lambda function
# lambda arguments: exp of what we want
a=lambda x: x**2
a(8)
#lambda function doesn't have a return type 
```

```python
su=lambda a,b: a+b
su(3,5)
```

```python
x=['kaist','gist','unistian','bindhu is awesome']#it basically sorts according to the alphabetical order
sorted(x) 
```

```python
# to sort based on the length
le = sorted(x,key=lambda l:len(x))
le
```

```python
fib = lambda n: n if n<=1 else fib(n-1)+fib(n-2)
[fib(i) for i in range(1,10)]
```

```python
#factorial of a number using lambda function
fact = lambda x: 1 if x<=1 else fact(x-1)*x
fact(5)
```

```python
# map syntax = map( function , iterable data structure)
l=[2,3,4,5,6]
def sq(i):
    return i**2
list(map (sq,l))

```

```python
# we can combine lambda and map as well
list(map(lambda i:i**2,l))
```

```python
# adding 2 lists values using lambda and map
l1=[1,2,3,4]
l2=[100,200,300,400]
list(map(lambda i,j:i+j,l1,l2))
```

```python
#to convert a string into upper case using lambda
s='kaist'
list(map(lambda i:i.upper(),s))
```

```python
#changing grades to cgpa
l=['A','B','C','D']
list(map(lambda i:4 if i=='A'else 2   ,l  ))
```

```python
from functools import reduce
```

```python
#syntax of reduce = reduce ( function , iterable data structure)
l=[1,2,3,4,5]
reduce(lambda x,y:x+y,l)
```

```python
words=['bindhu','got into','kaist']
reduce(lambda x,y:x+' '+y,words)
```

```python
n=[200,20,4,333,23,2233]
reduce(lambda x,y:x if x>=y else y,n)
```

```python
#filter function syntax = filter(function,iterable data structure)
#filter out the odd numbers
l=[2,1,3,4,6,4,32,45,66]
list(filter(lambda x:x%2==0,l))
```

```python
#filter out strings that are len 4 or more
p=['sd','sfjnksns','jiisj']
list(filter(lambda x:len(x)<4,p))
```
