```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('bi.csv',encoding='latin1')
data.head()
```

```python
#Data Cleaning
#checking the missing values
data.isnull().sum() #gives the sum of null values in each col
```

```python
#since there are less null values we can drop those rows
df = data.dropna()
```

```python
df.isnull().sum()
```

```python
#to get the datatype of each col
df.info() #df.dtypes
```

```python
#to check if there are any duplicate rows in a col
df.duplicated().sum()
```

```python
df.head()
```

```python
#Handling inconsistency in Gender col
df['gender'].value_counts()
```

```python
#Normalizing the gender col
df['gender']=(df['gender'].astype(str).str.strip().str.lower())

gender_map={'M':'male',
           'male':'male',
           'female':'female'}
df['gender']=df['gender'].map(gender_map)
df['gender'].value_counts()
```

```python
#Handling inconsistency in country col
df.country.value_counts()
```

```python
#Normalize and map
df['country']=(df['country'].astype(str).str.strip().str.lower()) #strip only removes the front and back spaces
df['country'] = df['country'].str.title()
df['country']=df['country'].replace({'Norge':'Norway','Somali':'Somalia'})
df['country'].value_counts()
```

```python
#Handling inconsistensies in prevEducation col
df['prevEducation'].value_counts()
```

```python
#Normalize and replace
df['prevEducation']=(df['prevEducation'].astype(str).str.strip().str.lower())
df['prevEducation']=df['prevEducation'].str.title()
df['prevEducation'].value_counts()
```

```python
df['prevEducation']=df['prevEducation'].replace({'Diplomaaa':'Diploma',
                                                'Highschool' : 'High School',
                                                'Barrrchelors' :'Bachelors'})
df['prevEducation'].value_counts()
```
