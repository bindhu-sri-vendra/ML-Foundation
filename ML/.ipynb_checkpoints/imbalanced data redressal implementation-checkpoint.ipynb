```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(1) # For reproducability of result
no_samples=1000
class_0_ratio=0.9
no_class0=int(no_samples*class_0_ratio)
no_class1=100
```

```python
no_class1,no_class0
```

```python
class_0={'feature1':np.random.normal(0,1,no_class0),
        'feature2':np.random.normal(0,1,no_class0),
        'target':[0]*no_class0}
```

```python
class_0=pd.DataFrame(class_0)
class_0
```

```python
class_1={'feature1':np.random.normal(3,1,no_class1),
        'feature2':np.random.normal(3,1,no_class1),
        'target':[1]*no_class1}
```

```python
class_1=pd.DataFrame(class_1)
class_1
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df
```

```python
df['target'].value_counts()
```

```python
#upsampling/over sampling >> we repeat data to make the classes equal
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_minority
```

```python
df_majority
```

```python
pip install scikit-learn
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsample= resample(df_minority,replace=True,n_samples=len(df_majority),random_state=1)
```

```python
df_minority_upsample
```

```python
df_upsample=pd.concat([df_majority,df_minority_upsample])
```

```python
df_upsample.target.value_counts()
```

```python
#down sampling / under sampling >> here we randomly select rows of upper class to match with lower class
df_downsample=resample(df_majority,replace=False,n_samples=len(df_minority),random_state=1)
df_downsample
```

```python
df_downsample.target.value_counts()
```

```python
df_down=pd.concat([df_minority,df_downsample])
df_down
```

```python
#SMOTE 
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[0.90])#n_redundant means we dont want to repeat the data
```

```python
x
```

```python
y
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
```

```python
final_df=pd.concat([df1,df2],axis=True)
```

```python
final_df
```

```python
final_df.target.value_counts()
```
