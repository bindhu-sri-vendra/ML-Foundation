```python
#data encoding basically means that we convert categorical data into numerical data
# nominal encoding ... binary vector for each of the category
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({'status':['single','married','separated','single','single','married','separated']})
df
```

```python
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[['status']]).toarray()
encoded
```

```python
encoder.get_feature_names_out()
```

```python
#to create a matrix structure for the dummy variables
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoder_df
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
#ordinal encoding .. asings unique values to each category based on a particular order or priority
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({'qualification':['HS','inter','Bachelors','masters','phd','HS']})
df
```

```python
encoder=OrdinalEncoder(categories=[['HS','inter','Bachelors','masters','phd']])
```

```python
encoder.fit_transform(df[['qualification']])
```
