```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
salary=[10,1,4,1,7,22,24,32,45,56,54,67,66,6,8,78,87,29,100,1000]
```

```python
#five point summary
np.quantile(salary,[0,0.25,0.50,0.75,1])
```

```python
df=pd.DataFrame(salary,columns=['salary'])
df
```

```python
df.describe() # we get the five point summary this way too
```

```python
#to check outliers
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['salary'],kde=True)
plt.title('Dist plot')

plt.subplot(1,2,2)# first row 2nd col 
sns.boxplot(data=df,x='salary')
plt.title('box plot')
plt.show()
```

```python
#dropping the outlier
#capping the outlier
#replacing with mean and median
```

```python
#dropping the outlier
Q1=df['salary'].quantile(0.25)
Q3=df['salary'].quantile(0.75)

IQR = Q3-Q1

lower_fence= Q1-1.5*IQR
upper_fence= Q3+1.5*IQR
```

```python
df.shape
```

```python
# the values that are below lower fence and above upper fence are called outliers so we drop them
df_filtered=df[(df['salary']>=lower_fence) & (df['salary']<=upper_fence)]
```

```python
df_filtered.shape # there is one outlier and that is dropped
```

```python
#imputation with mean and median
df['salary_imputed_mean']=np.where((df.salary>=upper_fence)|(df.salary<=lower_fence),df['salary'].mean(),df['salary'])
```

```python
df
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['salary_imputed_mean'],kde=True)
plt.title('Dist plot')

plt.subplot(1,2,2)
sns.boxplot(data=df,x='salary_imputed_mean')
plt.title('box plot')
plt.show()
```

```python
#capping the outlier >> means replacing the outlier with the nearest value that is not an outlier
lower_cap=df['salary'].quantile(0.05) # 5 percentile
upper_cap=df['salary'].quantile(0.95) # 95 percentile
```

```python
lower_cap
```

```python
upper_cap
```

```python
df['salary capped']=np.where(df['salary']<lower_cap,lower_cap,
        np.where(df['salary']>upper_cap,upper_cap,df['salary']))
```
