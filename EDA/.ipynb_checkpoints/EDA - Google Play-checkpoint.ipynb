```python
#lets keep the copy of original dataframe
df_copy=df.copy()
df_copy
```

```python
#since the row that's messing up reviews datatype is just one we can just drop it
df_copy[~df_copy.Reviews.str.isnumeric()] #the row after resetting the index
```

```python
#to drop a particular index
df_copy=df_copy.drop(df_copy.index[9990])
```

```python
df_copy
```

```python
#now we removed the row which messes up datatype of reviews we can update the datatype of the reviews
df_copy['Reviews']=df_copy['Reviews'].astype(int)
df_copy['Reviews'].dtype
```

```python
#now lets deal with another col
df['Size'].unique()
```

```python
#we know that 1mb=1024kb so we gonna convert all the m's in above data to kb by multiplying 1023 
def size_conversion(item):
    if item[-1] =='M':
        res=float(item.replace('M',''))#we gonna replace m with an empty string and then convert it into float 
        res*=1024
        return res
    elif item[-1] =='k':
        res=float(item.replace('k','')) 
        return res
    else:
        return np.nan
```

```python
size_conversion('heyyy')
```

```python
#now we apply the function on the col of our dataframe
df_copy['Size']=df_copy['Size'].apply(size_conversion)
```

```python
df_copy
```

```python
df_copy['Reviews'].dtype
```

```python
#we gotta handle the ,and + and Free
# we will replace them with an empty string
df_copy.Installs=df_copy.Installs.str.replace('+','').str.replace(',','')
```

```python
#now we gonna convert objects into strings
df_copy['Installs']=df_copy['Installs'].astype(int)
```

```python
#we can just replace the dollar with an empty string and then convert it into float
df_copy['Price']=df_copy['Price'].str.replace('$','')
df_copy['Price']=df_copy['Price'].astype(float)
```
