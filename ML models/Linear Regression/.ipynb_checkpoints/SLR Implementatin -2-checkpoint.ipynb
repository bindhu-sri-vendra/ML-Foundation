```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('height-weight.csv')
df
```

```python
#to check whether there is a linear relationship between height and weight --- we use scatter plot
plt.scatter(df.Weight,df.Height)
plt.xlabel('weight of the person')
plt.ylabel('height of the person')
plt.show()
```

```python
#problem Statement >>>> Predict the height based on weight of the person
```

```python
#Divide the data into x and y
x=df[['Weight']] # x is independent variable / can be multiple cols / used to predict the required thing
y=df['Height'] # y is dependant variable/ var that we are predicting
```

```python
# Train - Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
#scaling is optional >> standardization >> fit_transform is for train data and transform is for test data
#if we use fit on test data .. data leakage occurs .. its a blunder so avoid it
# Most ML models use standardScaler to deal with normal distribution
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
#model training
from sklearn.linear_model import LinearRegression
```

```python
#create an object for linear regression
regresor = LinearRegression()
regresor
```

```python
regresor.fit(x_train,y_train)
```

```python
regresor.coef_
#interpretation from coef_ .. on avg the height increases by 17.73 
```

```python
regresor.intercept_
```

```python
#prediction on test data
ypred = regresor.predict(x_test)
ypred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,ypred)
```

```python
mean_squared_error(y_test,ypred)
```

```python
r2_score(y_test,ypred)
```
