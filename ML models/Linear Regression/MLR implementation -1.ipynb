```python
# Multiple Linear Regression >> models the relationship between one dependent var y and multiple independent var's x
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the datasets
from sklearn.datasets import load_diabetes
diabetes= load_diabetes()
```

```python
diabetes.feature_names
```

```python
#convert the dataset into a dataframe
df=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
df
```

```python
df['target']=diabetes.target
df
```

```python
#EDA and data prep
X=df.drop('target',axis=1)# we are taking every single col except target as independent variables
y=df['target']
```

```python
#split the data into train and test data sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
#Model Training
from sklearn.linear_model import LinearRegression
regresor=LinearRegression()
regresor
```

```python
regresor.fit(X_train,y_train)
```

```python
regresor.coef_
```

```python
regresor.intercept_
```

```python
ypred=regresor.predict(X_test)
ypred
```

```python
# Model Evaluation
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(ypred,y_test)
```

```python
mean_absolute_error(ypred,y_test)
```

```python
r2_score(ypred,y_test)
```
