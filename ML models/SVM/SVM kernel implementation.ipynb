```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
x= np.linspace(-6,6,100) #100 values range from -6 to 6
x
```

```python
y=np.sqrt(10**2 - x**2)
y
```

```python
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

```python
x1 = np.linspace(-6,6,100) 
y1 = np.sqrt(4**2 - x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

```python
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
df1 = pd.DataFrame(np.vstack([y,x]).T,columns = ['x1','x2'])
df1['y']=0
df1
```

```python
df2 = pd.DataFrame(np.vstack([y1,x1]).T,columns = ['x1','x2'])
df2['y']=1
df2
```

```python
df = pd.concat([df1,df2], ignore_index = True)
df.head()
```

```python
df1 = df.copy()
df1 = df1.dropna()
```

```python
df1['x1square'] = df1['x1']**2
df1['x2square'] = df1['x2']**2
df1['x1x2'] = df1 ['x1']*df1['x2']
df1.head()
```

```python
X = df1[['x1square','x2square','x1x2']]
y=df1['y']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30, random_state=1)
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(df1,x='x1square',y='x2square',z='x1x2',color='y')
fig.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
cls = SVC(kernel='linear')
cls.fit(X_train,y_train)
y_pred = cls.predict(X_test)
accuracy_score(y_test,y_pred)
```
