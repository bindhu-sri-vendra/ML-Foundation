```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs #blobs are basically clusters
X,y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=1)
```

```python
X
```

```python
y
```

```python
#we only use X to find pattern in data in unsupervised learning
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
from sklearn.cluster import KMeans
kmeans= KMeans(n_clusters=3)
kmeans
```

```python
kmeans.fit(X_train) #fit_predict can also be used
```

```python
y_labels = kmeans.predict(X_train)
y_labels
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
```

```python
#how better this clustering model is >>> 
from sklearn.metrics import silhouette_score
silhouette_score(X_train,kmeans.labels_)#close to 1 means we have a good model
```

```python
kmeans.labels_
```

```python
kmeans.inertia_ # k=3 thsi is wcss distance
```

```python
#how to select k(centroid) >> elbow method
```

```python
wcss = []
for k in range(1,10):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
#the above result proves that wcss is decreasing as the k value increases
```

```python
#plot the elbow curve 
plt.plot(range(1,10),wcss)
plt.xticks(range(1,10))
plt.xlabel('no of students')
plt.ylabel('WCSS')
plt.show()
```

```python
#based on the plot above the optimal no of k/centroid should be 3
```
