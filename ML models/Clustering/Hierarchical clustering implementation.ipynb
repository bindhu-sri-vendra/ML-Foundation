```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
data = datasets.load_iris()
X=data.data
y=data.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#applying the pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(X_scaled)
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=y)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
#agglomerative clustering >> make dendograms

import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20,7))
plt.title('dendogram')

sc.dendrogram(sc.linkage(pca_scaled, method = 'ward'))

plt.xlabel('Sample data')
plt.ylabel('Euclidean distance')

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters =2, metric ='euclidean',linkage='ward')
cluster
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score

sil_coef = []
for k in range(2,11):
    agg = AgglomerativeClustering(n_clusters =k, metric='euclidean',linkage='ward')
    agg.fit(X_scaled)
    score = silhouette_score(X_scaled,agg.labels_)
    sil_coef.append(score)
```
