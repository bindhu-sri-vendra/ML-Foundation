```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
#problem statement >>> so basically there are 3 types of iris flowers and we are trying to predict which is which based on petal and sepal lenght and widths
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
data
```

```python
data.keys()
```

```python
print(data.DESCR)
```

```python
data.target
```

```python
data.feature_names
```

```python
#Making a dataframe
df=pd.DataFrame(data.data,columns=data.feature_names)
df['target']=data.target
df
```

```python
# if we wanna see what are the unique values from target col
df['target'].unique()
```

```python
#first we gonna practice binary classification so lets filter out the 2 from target col
df=df[df['target']!=2]
df
```

```python
df.target.unique()
```

```python
#Split the data into X and y
X=df.iloc[:,:-1] # every col except the last one 
y=df.iloc[:,-1] # only last col[target]
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
X_train.shape
```

```python
#model training
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model
```

```python
model.fit(X_train,y_train)
```

```python
ypred = model.predict(X_test)
ypred
```

```python
model.predict_proba(X_test)
```

```python
#to get predicted probabilities of class 1
model.predict_proba(X_test)[:,1]
```

```python
#Evaluation metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,ypred)
```

```python
accuracy_score(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
y_pred_proba = model.predict_proba(X_test)[:,1]
```

```python
#reciever operationg characteristics curve -roc
#inputs: ytest,ypred_proba. >> it will return TPR, FPR with diff cutoff probabilities

fpr,tpr, thresholds = roc_curve(y_test,y_pred_proba)
```

```python
thresholds
```

```python
tpr
```

```python
fpr
```

```python
roc_auc = auc(fpr,tpr)
```

```python
#plot roc-auc curve
plt.figure(figsize=(8,6)) 
plt.plot(fpr,tpr, color='darkorange',linewidth=2, label = 'ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0,1],[0,1],color='navy', linewidth=2, linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('Receiver Operating Characteristic (ROC) curve')
plt.legend(loc='lower right')
plt.show()
```
