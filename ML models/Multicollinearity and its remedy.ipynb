```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
data
```

```python
print(data.DESCR)
```

```python
# making the dataframe
df=pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
#adding the target col to the data frame
df['price']=data.target
df
```

```python
# multicolliniearity >>> when one feature is explained by more than or equal to 2 other features
# collinearity >>> when one feature is explained by only 2 features
```

```python
# now to figure out the correlation between features we use heatmap
sns.heatmap(df.corr(),annot=True,vmin=-1,vmax=1) #annot = True... displays corr in the heatmap directly , vmin and vmax are the range of heatmap
```

```python
#cluster map >>The rows and columns are automatically reordered to place features with similar correlation patterns next to each other, which helps you spot natural groupings in your data.
plt.figure(figsize=(3,4))
sns.clustermap(df.corr(),annot=True, vmin=-1,vmax=1)
plt.show()
```

```python
#problem with above map is we understand the relationship but we dont understand how much they are relation (quantitative measure)..so we study VIF
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['Feature']=df.columns # creating a datafrmae vif with col name features with values of all col names of california data set
```

```python
#to calculate the vif for all the features
vif['VIF']=[variance_inflation_factor(df.values,i) for i in range(len(df.columns))]
```

```python
vif
```

```python
#if vif is >10 we drop the features one by one
#make a copy and modify it
df1=df.copy()
df1
```

```python
#we drop the col with high vif and then calculate vif with other columns ... we repeat this process until all features have vif under 10
df1.drop('Longitude',axis=1,inplace=True)
```

```python
df1
```

```python
#now we calculate vif once again
vif=pd.DataFrame()
vif['Feature']=df1.columns
vif['VIF']=[variance_inflation_factor(df1.values,i) for i in range (len(df1.columns))]
vif
```

```python
#now we drop avg rooms cause it has the highest vif
df1.drop(columns='AveRooms',inplace=True)
df1
```

```python
#now we calculate vif once again
vif=pd.DataFrame()
vif['Feature']=df1.columns
vif['VIF']=[variance_inflation_factor(df1.values,i) for i in range (len(df1.columns))]
vif
```

```python
#now we drop avg rooms cause it has the highest vif
df1.drop(columns='Latitude',inplace=True)
df1
```

```python
vif=pd.DataFrame()
vif['Feature']=df1.columns
vif['VIF']=[variance_inflation_factor(df1.values,i) for i in range (len(df1.columns))]
vif
```

```python
#now all the VIF values are below 10 so we can build the model
X=df1.iloc[:,:-1] #independent var
y=df1.iloc[:,-1] # dependent var
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
X_train
```

```python
# we do scaling after train_test_split which is optional
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler
```

```python
#fit_transform is used for train data
#transform is used for test data 
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model
```

```python
#to train the model
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
#performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
