```python
#adaboost classifier

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score , classification_report, confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
X,y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=1)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=1)
```

```python
classifier= AdaBoostClassifier()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
ypred = classifier.predict(X_test)
```

```python
#Model Evaluation
print('current model performance')
print(f'accuracy: {accuracy_score(y_test, ypred)}')
print(classification_report(y_test, ypred))
print(confusion_matrix(y_test, ypred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators':[50,100,200],
             'learning_rate':[0.01,0.1,1.0,1.5,2.0],
             'algorithm':['SAMME']
             }
```

```python
ada = AdaBoostClassifier()
clf = GridSearchCV(estimator=ada,param_grid = param_grid, cv=5,verbose=3)
clf
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
best_model = clf.best_estimator_
```

```python
ypred = best_model.predict(X_test)

#Model Evaluation

print('current model performance')
print(f'accuracy: {accuracy_score(y_test, ypred)}')
print(classification_report(y_test, ypred))
print(confusion_matrix(y_test, ypred))
```

```python
#adaboost regressor

from sklearn.datasets import make_regression
from sklearn.ensemble import AdaBoostRegressor
```

```python
X , y = make_regression(n_samples=1000, n_features=2, noise=10, random_state=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=1)
```

```python
regressor = AdaBoostRegressor()
regressor
```

```python
regressor.fit(X_train, y_train)
```

```python
ypred = regressor.predict(X_test)
```

```python
#model evaluation

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(y_test,ypred))
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#hyperparameter tuning
param_grid = {'n_estimators':[50,100,200],
             'learning_rate':[0.01,0.1,1.0,1.5,2.0],
             'loss':['linear','square','exponential']
             }
```

```python
grid_search = GridSearchCV(estimator=regressor, param_grid=param_grid, cv=5,verbose=3)
grid_search
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```
