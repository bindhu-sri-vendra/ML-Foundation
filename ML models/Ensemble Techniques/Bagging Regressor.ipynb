```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

X,y = make_regression(n_samples=1000, n_features=10, noise=0.1, random_state=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.20, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
sv = SVR(kernel='linear')
```

```python
from sklearn.ensemble import VotingRegressor
ensemble_reg = VotingRegressor(estimators=[('linear reg',lr),('decision tree',dt),('svr',sv)])
```

```python
ensemble_reg
```

```python
ensemble_reg.fit(X_train,y_train)
```

```python
ypred = ensemble_reg.predict(X_test) #the final prediction will be the average of all predictions
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypred)
```
