```python
import numpy as np
```

```python
data=[1,2,3,4,5,23,22,24,12,13,17,67,56]
```

```python
#range = max - min
max(data)-min(data)
```

```python
#percentile 
np.percentile(data,[0,25,50,75,100])
```

```python
#IQR = Q3-Q1 = 75 percentile - 25 percentile .... it is used to remove the outlies
23-4
```

```python
#box plot
import seaborn as sns
sns.boxplot(data)
```

```python
#variance 
np.var(data)
```

```python
#standard deviation
np.std(data)
```
