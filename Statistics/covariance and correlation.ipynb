```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
#covariance and correlation are always calculated for numerical features
df.cov(numeric_only=True)
```

```python
#correlation is a dimension less quantity and has range -1 to 1
df.corr(numeric_only=True)
```

```python
cor=df.corr(numeric_only=True)
sns.heatmap(cor)
```

```python
sns.heatmap(cor,annot=True)
```
