```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
import plotly.io as pio
pio.renderers.default = 'notebook'
```

```python
#using graph objects first
#add_trace .... adding data 
fig=go.Figure()#to create a page for the plot
fig.add_trace(go.Scatter(x=[1,2,3,6,8],y=[3,8,5,1,3],mode='markers'))# we are working with a scatter plot
```

```python
#to create a line plot just update markers as line
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,6,8],y=[3,8,5,1,3],mode='lines'))
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers'))
```

```python
#bar plot
fig=go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5],y=[3,8,5,6,9]))
```

```python
#to make a histogram
fig=go.Figure()
fig.add_trace(go.Histogram(x=tips.total_bill))
```

```python
#lets say in scatter plot, we wanna see the dot bigger for bigger group of people at restaurant from tips data
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=10*tips['size']))
```

```python
#3d scatter plot
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=10*tips['size'],z=tips['size']))
```

```python
#3d line chart
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=[1,2,3,4,5],y=[2,6,4,5,6],z=[2,5,6,7,8]))
```

```python
#now we use plotly express to make our plots, it has an easier syntax compared to plotly graph object
x=[1,2,3,4,5]
y=[12,16,14,15,16]
fig=px.scatter(x,y) #px is the plotly express object
fig.show()
```

```python
#line plot in plotly express
px.line(x,y)
```

```python
#bar plot
px.bar(x,y)
```

```python
#histogram in plotly express
x=[1,2,5,3,4,6,7,5,7,9]
fig=px.histogram(x)
fig.show()
```

```python
x=['a','b','c']
y=[1,5,10,3,25,12,33,100]
fig=px.box(x,y)
fig.show()
```

```python
#plotting heat map
x=[[1,2,4],[3,5,6],[3,8,9]]
fig=px.imshow(x)
fig.show()
```

```python
y=[5,8,12,13]
x=[1,2,3,4]
fig=px.scatter(x,y,size=[30,10,50,100])
fig.show()
```

```python
#3d chart
y=[5,8,12,13]
x=[1,2,3,4]
z=[3,4,5,6]
px.scatter_3d(x=x,y=y,z=z)
```
