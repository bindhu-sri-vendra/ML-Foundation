```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv')
```

```python
#accessing rows and cols >> we have two ways implicit and explicit
df.iloc[0:3] # it is implict access of rows and it starts from 0 and continues till 2
#IMP >> it is used for integer indexes
```

```python
df.loc[0:3] # this is the explicit way of accessing and it starts from 0 and ends with 3
# IMP >> it is used for named indexes 
```

```python
# we cant access cols through iloc currently cause they are named cols so we need to use loc
df.loc[0:2 , ['Name','Age','Ticket']]
```

```python
#since we know the cols are indexed defaulty starting with 0 we can use iloc through it 
df.iloc[0:3,3:5] # 0 passid 1 survived 2 pclass 3 name 4 sex 5 age like that
```

```python
data={'key1':[1,2,None,3,4,5,None,8,9,10],
     'key2':[11,12,13,None,14,None,10,23,22,None]}
dfr=pd.DataFrame(data)
dfr
```

```python
dfr['key1'].mean()
```

```python
#to fill the missing values with the mean of the col
dfr['key1'].fillna(dfr['key1'].mean())
```

```python
#to fill the missing values with the median of the col
dfr['key1'].fillna(dfr['key1'].median())
```

```python
# ffill aka forward fill >> starting from the bottom of the col if we encounter a null val we will fill it by the above val 
dfr.fillna(method='ffill')
```

```python
dfr
```

```python
dfr.ffill()
```

```python
# we also have bfill it works in the opposite way of ffill
dfr.bfill()
```

```python
#to check any duplicated rows are there or not
dfr.duplicated().sum()
```
