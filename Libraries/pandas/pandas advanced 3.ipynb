```python
dd1=pd.DataFrame({'key1':[1,2,3,4,5,6],
                 'key2':[2,6,4,6,7,3],
                 'key3':[2,7,8,9,5,3]})
dd2=pd.DataFrame({'key1':[8,2,3,9,5,6],
                 'key4':[2,9,4,6,5,3],
                 'key5':[4,7,8,9,5,3]})
```

```python
pd.merge(dd1,dd2,how='inner') # by default is inner merge 
```

```python
pd.merge(dd1,dd2 ,how='left')
```

```python
pd.merge(dd1,dd2 ,how='right')
```

```python
#join works based on indexes and merge works based on cols
dd1=pd.DataFrame({'key1':[1,2,3,4,5,6],
                 'key2':[2,6,4,6,7,3],
                 'key3':[2,7,8,9,5,3]},index=['a','b','c','d','e','f'])
dd2=pd.DataFrame({'key6':[8,2,3,9,5,6],
                 'key4':[2,9,4,6,5,3],
                 'key5':[4,7,8,9,5,3]},index=['a','b','h','i','j','k'])
```

```python
dd1.join(dd2,how='left')
```

```python
# we can also use the user defined function with apply 
def convert(x):
    return x*100
df['Fare'].apply(convert) # we just need to pass the name of the function
```

```python
#Q make a new col that says whether the fare is low or medium or high
def cate(x):
    if x<10:
        return 'low fare'
    elif x>=10 and x<=20:
        return 'medium fare'
    else:
        return 'high fare'
df['fare category']=df['Fare'].apply(cate)
df
```
