```python
#to check the execution time of the code/program
start=pd.Timestamp.now()
for i in range(1000):
    pass
end=pd.Timestamp.now()
print('the execution time of the program is ',end-start)
```

```python
#another way to get the execution time of the program ... it gives the answer in seconds
import time as t
start=t.time()
for i in range (1000):
    pass
end=t.time()
print('the execution time of the program is ',end-start)
```

```python
#another way >>>>> prefer this ....industry standards
start=t.perf_counter()
for i in range (1000):
    pass
end=t.perf_counter()
print('the execution time of the program is ',end-start)
```

```python
#to know the date and time a week before from now exactly
pd.Timestamp.now()-pd.DateOffset(weeks=1)
```
