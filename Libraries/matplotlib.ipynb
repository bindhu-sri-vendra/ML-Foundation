```python
#lineplot >> used when we want to understand the trend between two variable or cols or functions
x=[1,2,3,4,5]
y=[5,2,7,8,1]
plt.plot(x,y,color='pink')
plt.title('line chart to understand two lists')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show() # to remove "Text(0, 0.5, 'price as y points')" at the top of the figure and just display the chart
```

```python
plt.plot(x,y,color='pink',marker='o')
plt.title('line chart to understand two lists')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show()
```

```python
plt.plot(x,y,color='pink',marker='o',linestyle='-.')#to change the style of the line >>>> supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.title('line chart to understand two lists')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show()
```

```python
#to increase the width of the line
plt.plot(x,y,color='pink',marker='o',linestyle='-.',linewidth=3)
plt.title('line chart to understand two lists')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show()
```

```python
plt.plot(x,y,color='pink',marker='o',linestyle='-.',linewidth=3,markersize=10) #marker is used to make those plotted circles big
plt.title('line chart to understand two lists')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show()
```

```python
x=np.linspace(1,10,100) #it generates 100 points between 1 and 10
```

```python
# we can plot with mathematical functions as well
plt.plot(x,np.sin(x),color='pink')
plt.plot(x,np.cos(x))
```

```python
#to make the plot side by side 
plt.figure()# initializing a page to make the plot
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))
plt.subplot(1,2,2)
plt.plot(x,np.cos(x))
```

```python
plt.figure()
plt.subplot(2,1,1)# it says 2 rows and 1 col and it is the first plot
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)# it says  it is the second plot
plt.plot(x,np.cos(x))
```

```python
# we can also put limits on values that are on x and y axis
plt.plot(x,np.sin(x))
plt.xlim(1,15)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
#the graph is stopped in the middle we can fix that
plt.plot(x,np.sin(x))
plt.xlim(1,15)
plt.ylim(-1.5,1.5)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
#histogram >> it takes continous data (any number on the number line) --- basically when we want to understand continuous data we use histogram
data=np.random.randn(100)
plt.hist(data)
plt.grid()
plt.show()
```

```python
plt.hist(data,color='lightgreen',bins=15) # we can increase the no of bins we want and can change the color of the histograph
plt.grid()
plt.show()
```

```python
#stacked histogram >> two histograms together
data1=np.random.normal(0,1,5000)
data2=np.random.normal(3,2,5000)
```

```python
plt.hist([data1,data2],bins=15,color=['green','black'],label=['dist1','dist2'])
plt.title('stacked histogram example')
plt.xlabel('the x values')
plt.ylabel('the y values')
plt.legend()
plt.show()
```

```python
#pie chart is used to understand the percent distribution of the data
lang=['c','cpp','python']
students=[12,45,65]# no of students in each lang
plt.pie(students,labels=lang)
plt.show()
```

```python
plt.pie(students,labels=lang ,autopct='%1.2f%%',shadow=True) #autopct is used to get the percentage in pie chart and 2f means we need upto 2 decimal points
plt.show()
```

```python
#to foucs on one segment
plt.pie(students,labels=lang ,autopct='%1.2f%%',explode=(0.0,0.1,0.1)) 
plt.show()
```

```python
#we can also provide shadow to our pie chart
plt.pie(students,labels=lang ,autopct='%1.2f%%',explode=(0.0,0.1,0.1),shadow=True) 
plt.show()
```

```python
#now applying the plots in real world data 
df=pd.read_csv('Bank_churn.csv')
df
```

```python
#we can also write the above plot in different way
df.plot.scatter(x='Age',y='Balance')
```

```python
#1,2,3,4,100 here 100 is abnormal and called as outlier(extreme data) .. we gotta remove them its imp so we can use box plot
df.boxplot(column='Balance')
```

```python
years=[2020,2025,2015,2023,2007]
category1=[23,12,43,54,61]
category2=[12,17,23,33,54]
category3=[3,2,4,5,1]
```

```python
plt.stackplot(years,category1,category2,category3,labels=['category1','category2','category3'],colors=['black','pink','skyblue'])
plt.title('Area plot')
plt.xlabel('years')
plt.ylabel('values')
plt.legend()
plt.grid()
plt.show()
```
