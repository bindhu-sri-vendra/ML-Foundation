```python
l=[1,2,3,4,5]
arr=np.array(l)
```

```python
#accessing and updating elements in an array
arr[0]=100
arr
```

```python
a=arr# here both the arrays point to the same memory location so changes in the one array reflects in the second
a[1]=20
arr
```

```python
#if we want just a copy of an array then we use copy()
b=arr.copy()
b[2]=30 # we changed b but the change wont reflect in arr
arr
```
