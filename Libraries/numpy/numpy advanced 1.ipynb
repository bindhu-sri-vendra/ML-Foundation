```python
a=np.random.randint(1,5,(3,4)) # generates an array of size 3,4 with the elements between 1 and 5
a
```

```python
#to reshape an array
a.reshape(6,2)
```

```python
#in reshaping just giving no of rows also works
a.reshape(4,-1)#4 is the no of rows i desire
```

```python
#similar with cols as well
a.reshape(-1,3) #we desire 3 cols and we can pass any negative value in place of -1 and it works the same
```

```python
#after reshaping if we wanna see the original array then we can use base function
a.reshape(1,12).base
```

```python
#mathematical operations on arrays
arr1=np.random.randint(1,4,(3,3))#generates a 3x3 array with the values between one and four
arr2=np.random.randint(1,5,(3,3))
arr1
```

```python
arr2
```

```python
#addition of 2 arrays 
arr1 + arr2
```

```python
#multiplication of 2 arrays >> its different from matrix multiplication
arr1 * arr2
```

```python
#div of 2 arrays
arr1/arr2
```

```python
#matrix multiplication of two arrays
arr1 @ arr2
```

```python
#alternate way to matrix multiplication
np.dot(arr1,arr2)
```
