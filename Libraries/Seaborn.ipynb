```python
#to know the available datasets in seaborn
sns.get_dataset_names()
```

```python
#to load our desired dataset from seaborn
df=sns.load_dataset('tips')
df
```

```python
#numerical data plotting
#seaborn's relplot(relationship plot) has the option to plot both numerical and categorical data
#relplot is used basically to understand the relation between 2 variables/features/cols
#by default relplot makes scatter plot
```

```python
#we wanna see how tip is varying with respect to total_bill
sns.relplot(x='total_bill',y='tip',data=df)
#insight >> tips increased with respective to total bills
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker')
#insight >> smoking people gave high tips
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='size')#hue gives color coded options and sometimes they are hard to understand when there are too many
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='size')#now it gives diff styles to diff categories
```

```python
#we can also use style and hue at the same time 
sns.relplot(x='total_bill',y='tip',data=df,style='smoker',hue='time')
#insights >> orange x means he came for dinner and is a non smoker >> blue dot means he came for lunch and is a smoker
```

```python
#we can also use relplot for making the line plot
#lineplot is used to understand the trend of the data and usually time col/sequence will be there in the data
time=np.arange(500)
value=np.random.randn(500).cumsum()
df=pd.DataFrame({'time':time,'value':value})
df
```

```python
sns.relplot(x='time',y='value',data=df,kind='line',sort=True)
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#two subplots based on columns
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',col='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',col='size')
```

```python
#categorical plots
tips
```

```python
sns.catplot(x='day',y='tip',data=tips)
```

```python
#insights for above graph >> more tips came on weekends especially saturday
#in the above plot some values are overlapping on each other so we use swarmplot to view perfectly
sns.catplot(x='day',y='tip',data=tips,kind='swarm')
```

```python
#boxplot to see the outliers
sns.catplot(x='day',y='tip',data=tips,kind='box')
#the values above the top line are outliers/extreme values which need to be eliminated
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='violin')
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='boxen')#the diff between box and boxen is the no of levels
```

```python
#barplot
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
#bivariable plot
sns.jointplot(x=tips.total_bill,y=tips.tip)#draws a plot of two variables with distribution
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip,kind='reg')#reg means regression line
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip,kind='hex')
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip,kind='kde')#kernel density estimator >> used in metrological department
```

```python
sns.pairplot(tips)#it makes scatter plots with respective of all numerical cols
```

```python
sns.pairplot(tips,hue='smoker')
```

```python
sns.pairplot(tips,hue='smoker',palette='coolwarm')
```
