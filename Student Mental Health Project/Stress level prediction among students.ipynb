```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('students_mental_health_survey.csv')
data.head()
```

```python
data.CGPA.unique()
```

```python
sns.histplot(data['CGPA'].dropna(),kde=True)
```

```python
data['CGPA'] = data['CGPA'].fillna(data['CGPA'].median())
```

```python
data['Substance_Use'].value_counts()
```

#since 'Substance_Use' is a categorical column we replace the missing vales with mode 

```python
data['Substance_Use'] = data['Substance_Use'].fillna('Never')
```

```python
data.dtypes
```

```python
data.Stress_Level.unique()
```

```python
data.Stress_Level.hist()
```

```python
#Separating numeric and categorical columns 
num_col = [col for col in data if data[col].dtype !='O']
cat_col = [col for col in data if data[col].dtype == 'O']
```

Univariate Analysis of Numeric Columns

```python
for col in num_col:
    plt.figure()
    data[col].hist(bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()
```

Univariate Analysis of Categorical Columns 

```python
for col in cat_col:
    data[col].value_counts().plot(kind = 'bar')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()
```

Bivariate Analysis of Target Col vs Numeric columns

```python
for col in num_col:
    if col!='Stress_Level':
        plt.figure()
        sns.scatterplot(x = data[col], y = data['Stress_Level'])
        plt.title(f'{col} vs Stress level')
        plt.xlabel(col)
        plt.ylabel('Stress level')
        plt.show()
```
